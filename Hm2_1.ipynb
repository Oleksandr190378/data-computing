```python
import pandas as pd
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
data = pd.read_html(
    url,
    match="Коефіцієнт народжуваності в регіонах України",
    decimal=",",
    thousands=".",
)[0]
```

```python
data.head()
```

```python
data.shape
```

```python
data[['2014', '2019']] = data[['2014', '2019']].apply(pd.to_numeric, errors='coerce')
```

```python
data.dtypes
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop([len(data)-1], inplace=True)
```

```python
import numpy as np
numeric_columns = data.select_dtypes(include=[np.number]).columns
data[numeric_columns] = data[numeric_columns].transform(lambda col: col.fillna(col.mean()).round(1), axis=0)
data
```

```python
regions =  data[data['2019'] > data['2019'].mean()]['Регіон'].tolist()
regions
```

```python
region = data[data['2014'] == data['2014'].max()]['Регіон'].tolist()
region
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,8))
plt.bar(data['Регіон'], data['2019'], color='b')
plt.xlabel("Регіон", fontsize="small", color="midnightblue")
plt.xticks(rotation=45)
plt.ylabel("Кількість", fontsize="small", color="midnightblue")
plt.title("Рівень народжуваності по регіонах у 2019 році", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

hm_data = data[data['Регіон'] == 'Хмельницька']
print(hm_data)
plt.plot(hm_data.columns[1:], hm_data.values[0, 1:], color='b', label="Рівень народжуваності")
plt.xlabel("Pik", fontsize="small", color="midnightblue")
plt.xticks(rotation=45)
plt.ylabel("Рівень народжуваності ", fontsize="small", color="midnightblue")
plt.title("Рівень народжуваності в Хмельницькій області", fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

hm_data = data[data['Регіон'] == 'Київ']
plt.plot(hm_data.columns[1:], hm_data.values[0, 1:], linestyle="--",
    color="#FF5733",label="Рівень народжуваності")
plt.xlabel("Pik", fontsize="small", color="midnightblue")
plt.xticks(rotation=45)
plt.ylabel("Рівень народжуваності ", fontsize="small", color="midnightblue")
plt.title("Рівень народжуваності в Києві", fontsize=15)
plt.legend()
plt.grid()
plt.show()
```
